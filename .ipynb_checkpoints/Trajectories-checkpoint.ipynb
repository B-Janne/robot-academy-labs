# Paths and Trajectories
## 1D polynomial trajectory

1D polynomial trajectories are commonly used to define smooth paths for moving objects. These trajectories are defined by polynomials in time, and their coefficients are determined based on boundary conditions (e.g., start and end positions, velocities, and accelerations).

### Polynomial Representation:
A 1D polynomial trajectory is represented as:
$$
x(t) = a_n t^n + a_{n-1} t^{n-1} + \dots + a_1 t + a_0
$$
Where:
- $ x(t) $ is the position at time $ t $.
- $ a_n, a_{n-1}, \dots, a_0 $ are the polynomial coefficients.
- $ n $ is the order of the polynomial.
   
The order of the polynomial depends on the number of boundary conditions. For example, Aa 5th-order polynomial can satisfy 6 boundary conditions (position, velocity, and acceleration at the start and end).

```python
import numpy as np
import matplotlib.pyplot as plt
```

We define the start and end times and positions. We also set up a system of equations to solve for the polynomial coefficients which are encoded in the matrix A and the vector b.

```python
# Define the start and end times
t_start = 0
t_end = 10

# Define the start and end positions
p_start = 0
p_end = 10

# Define the polynomial coefficients
A = np.array([
    [0, 0, 0, 0, 0, 1],  # x(0) = p_start
    [t_end**5, t_end**4, t_end**3, t_end**2, t_end, 1],  # x(t_end) = p_end
    [0, 0, 0, 0, 1, 0],  # dx/dt(0) = 0
    [5*t_end**4, 4*t_end**3, 3*t_end**2, 2*t_end, 1, 0],  # dx/dt(t_end) = 0
    [0, 0, 0, 2, 0, 0],  # d^2x/dt^2(0) = 0
    [20*t_end**3, 12*t_end**2, 6*t_end, 2, 0, 0]  # d^2x/dt^2(t_end) = 0
])
b = np.array([p_start, p_end, 0, 0, 0, 0])
```

We compute the position, velocity, and acceleration over time using the polynomial coefficients and then plot the result.

```python
# Solve for the coefficients 
if np.linalg.det(A) != 0:
    coeffs = np.linalg.inv(A) @ b
else:
    raise ValueError("Cannot find polynomial coefficients.")

# Define the time vector
steps = 100
t = np.linspace(t_start, t_end, steps)

# Calculate the position vector
x = np.polyval(coeffs, t)

# Compute the velocity and acceleration
dx = np.polyval(np.polyder(coeffs), t)  # First derivative
ddx = np.polyval(np.polyder(coeffs, 2), t)  # Second derivative

plt.style.use('ggplot')

# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(6, 6), sharex=True)

# Position plot
axs[0].plot(t, x, color='red', linewidth=0.5)
axs[0].set_ylabel('Position (m)')
axs[0].set_title('1D Polynomial Trajectory')

# Velocity plot
axs[1].plot(t, dx, color='red', linewidth=0.5)
axs[1].set_ylabel('Velocity (m/s)')

# Acceleration plot
axs[2].plot(t, ddx, color='red', linewidth=0.5)
axs[2].set_xlabel('Time (s)')
axs[2].set_ylabel('Acceleration (m/s²)')

# Adjust layout for better readability
plt.tight_layout()
plt.show()
```

## 1D trapezoidal trajectory

Polynomial trajectories are often used to ensure smooth motion. However, the object spends significant time accelerating and decelerating, meaning it moves slower than it could for much of the trajectory. 

To improve efficiency, a trapezoidal trajectory can be used. This trajectory maximizes velocity utilization by accelerating quickly to a desired speed, maintaining that speed for as long as possible, and then decelerating smoothly to stop at the final position. It consists of three phases:
1. **Acceleration Phase**: The object accelerates from rest to a maximum velocity.
2. **Constant Velocity Phase**: The object moves at a constant maximum velocity.
3. **Deceleration Phase**: The object decelerates to a stop.

```python
from roboticstoolbox import trapezoidal

# Define the start and end positions
p_start = 0
p_end = 10

# Define time steps
steps = 10

# Generate the trapezoidal trajectory
traj = trapezoidal(p_start, p_end, steps)
traj.plot()
```

## Trajectory Through a Series of Points with Blends
Moving through a series of discrete points is common. However, abrupt changes in velocity at each waypoint can lead to inefficient or physically infeasible trajectories. Instead of stopping at each via-point, we use blended trajectories, where smooth transitions are generated between waypoints while maintaining continuous velocity and acceleration constraints.

```python
from roboticstoolbox import mstraj 

# Define time step
dt = 2  

# Define via-points (X, Y) to be followed
viapoints = np.array([
    [60, 30], 
    [40, 10], 
    [20, 30], 
    [40, 50]
])

# Define acceleration time and velocity constraints
tacc = 0.1  # Acceleration time (smooth transition)
SPEED = [1, 1]  # Maximum velocity for each dimension (x, y)

# Compute the trajectory through via-points with blending
traj = mstraj(viapoints, dt, tacc, qdmax=SPEED, q0=[60, 30])

# Plot via-points as red circles
plt.plot(viapoints[:, 0], viapoints[:, 1], 'o-r', markersize=3, label="Via-points")

# Plot interpolated trajectory as blue dots
plt.plot(traj.q[:, 0], traj.q[:, 1], 'o-b', markersize=3, label="Blended Trajectory")

plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.title("Multi-Segment Blended Trajectory")
plt.legend()
plt.show()
```
